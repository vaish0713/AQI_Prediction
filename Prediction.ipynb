```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
```

```python
df=pd.read_csv("city_day.csv")
df.head()
```

```python
import math
med_pm25 = math.floor(df["PM2.5"].median())
df["PM2.5"]=df["PM2.5"].fillna(med_pm25)
df.head()
```

```python
import math
df_PM10=math.floor(df.PM10.median())
df.PM10=df.PM10.fillna(df_PM10)
df.head()
```

```python
import math
df_NH3=math.floor(df.NH3.median())
df.NH3=df.NH3.fillna(df_NH3)
df.head()
```

```python
df_AQI=math.floor(df.AQI.median())
df.AQI=df.AQI.fillna(df_AQI)
df.head()
```

```python
df_NO=math.floor(df.NO.median())
df.NO=df.NO.fillna(df_NO)
df.head()
```

```python
df_NO2=math.floor(df.NO2.median())
df.NO2=df.NO2.fillna(df_NO2)
df_NOx=math.floor(df.NOx.median())
df.NOx=df.NOx.fillna(df_NOx)
df_CO=math.floor(df.CO.median())
df.CO=df.CO.fillna(df_CO)
df_SO2=math.floor(df.SO2.median())
df.SO2=df.SO2.fillna(df_SO2)
df_O3=math.floor(df.O3.median())
df.O3=df.O3.fillna(df_O3)
df_Benzene=math.floor(df.Benzene.median())
df.Benzene=df.Benzene.fillna(df_Benzene)
df_Toluene=math.floor(df.Toluene.median())
df.Toluene=df.Toluene.fillna(df_Toluene)
df_Xylene=math.floor(df.Xylene.median())
df.Xylene=df.Xylene.fillna(df_Xylene)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']],df['AQI'])
```

```python
var=reg.predict([[10, 20, 30, 40, 50, 20, 1, 15, 35, 3, 25, 15]])
```

```python
if(var>300):
    print("Harzardous")
elif(var<300 and var>200):
  print("Very unhealthy")
elif(var>151 and var<200):
  print("Unhealthy")
elif(var>101 and var<150):
  print("Unhealthy for certain groups")
elif(var>50 and var<100):
  print("Moderate")
elif(var>0 and var<50):
  print("Good")
```
